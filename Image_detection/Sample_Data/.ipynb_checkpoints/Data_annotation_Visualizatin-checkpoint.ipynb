### Notes

- Filenames should be integers
    - 0.jpg
    - 93535_3534534_343345.jpg
    - 4530954_354564.png

- Download in voc xml format here: https://github.com/Tony607/voc2coco
- convert to COCO json using voc2coco.py 

- coco.anns
    - Gives no of bboxes in all images.
- coco.imgs
    - Gives no of images present
    
##### No of images != No of bboxes

```python
import os
import pandas as pd
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt

from PIL import Image, ImageDraw, ImageFont
from pycocotools.coco import COCO
%matplotlib inline
```

```python
coco = COCO("/home/roufa/Desktop/interview/pytorch/Image_detection/Sample_Data/2.coco_annotations/output.json")
```

```python
# display COCO categories and supercategories
cats = coco.loadCats(coco.getCatIds())
nms=[cat['name'] for cat in cats]
print('COCO categories: \n{}\n'.format(' '.join(nms)))
```

```python
coco.imgs
```

```python
# get all images containing given categories, select one at random
catIds = coco.getCatIds(catNms=["Ants","Bees"]);
imgIds = coco.getImgIds(catIds=catIds );
imgIds = coco.getImgIds(imgIds = [4])
img = coco.loadImgs(imgIds[np.random.randint(0,len(imgIds))])[0]
```

```python
I = Image.open("/home/roufa/Desktop/interview/pytorch/Image_detection/Sample_Data/images/004.jpg")
plt.axis('off')
plt.imshow(I)
plt.show()
```

```python
# load and display instance annotations
annIds = coco.getAnnIds(imgIds=img['id'], catIds=catIds, iscrowd=None)
anns = coco.loadAnns(annIds)

def plot(anns, I):
    x1,y1,w,h = anns[0]['bbox']
    x1,y1,x2,y2 = x1, y1, x1+w, y1+h
    
    draw = ImageDraw.Draw(I)
    draw.rectangle([x1,y1,x2,y2], fill=None, outline=None, width=5)
    plt.imshow(I)
    plt.show()

plot(anns, I)
```

### FOr all

```python
def plot1(anns, I, label, font):
    x1,y1,w,h = anns['bbox']
    x1,y1,x2,y2 = x1, y1, x1+w, y1+h
    
    draw = ImageDraw.Draw(I)
    draw.rectangle([x1,y1,x2,y2], fill=None, outline=None, width=5)
    draw.text((x1,y1), label, fill=(255, 0, 0), font=font)
    
    return I
```

```python
root_path = "/home/roufa/Desktop/interview/pytorch/Image_detection/Sample_Data/images/"
labels = ["ants","bees"]
font = ImageFont.truetype("Aaargh_font/Aaargh.ttf", size = 30)

for index in range(len(coco.imgs)):
    catIds = coco.getCatIds(catNms=["Ants","Bees"]);    
    imgIds = coco.getImgIds(imgIds = [index+1])
    img = coco.loadImgs(imgIds)[0]
    image = Image.open(os.path.join(root_path,img['file_name']))    
    
    annIds = coco.getAnnIds(imgIds=img['id'], catIds=catIds, iscrowd=None)    
    anns = coco.loadAnns(annIds)
    
    for ann in anns:
        image = plot1(ann, image, labels[ann['category_id']], font)
    plt.imshow(image)
    plt.show()
```
